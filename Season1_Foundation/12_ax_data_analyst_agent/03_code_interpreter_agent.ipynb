# 🚀 Code Interpreter: 데이터 시각화 및 자가 수정(Self-Correction) 에이전트

```python
import traceback
import sys
import pandas as pd
import matplotlib.pyplot as plt

# 이전 단계에서 추출된 DataFrame (df_result)이 있다고 가정합니다.
# 시뮬레이션을 위해 엔진 1번의 5사이클치 데이터 생성
data = {'cycle': [1,2,3,4,5], 'temperature': [641.82, 642.15, 642.35, 642.35, 642.37]}
df_result = pd.DataFrame(data)

print("👨‍💼 [지시사항]: '추출된 데이터를 바탕으로, x축이 cycle이고 y축이 온도인 꺾은선 그래프를 그려서 저장해줘.'\n")

# --- 에러가 발생하는 LLM의 1차 코드 ---
llm_code_attempt_1 = """
plt.figure(figsize=(10,5))
plt.plot(df_result['cycle'], df_result['temperatur'], color='red') # 의도적인 오타: temperature -> temperatur
plt.title("Engine 1 Temperature over Cycles")
plt.xlabel("Cycle")
plt.ylabel("Temp")
plt.savefig("engine1_report.png")
"""
print("🤖 [LLM 작성 코드 (1차)]:")
print(llm_code_attempt_1)
```

```python
print("⚙️ [에이전트 엔진]: 로컬 샌드박스에서 코드를 실행합니다...\n")
error_log = ""
try:
    exec(llm_code_attempt_1)
except Exception as e:
    error_log = traceback.format_exc()
    print("❌ [실행 에러!] 에러 로그를 캡처하여 LLM에게 재수정을 요청합니다.")
    print("Error:", error_log.strip().split('\n')[-1])
```

```python
llm_thought = """Observation: `KeyError: 'temperatur'` 발생.
Thought: df_result의 컬럼명 철자가 틀렸다. 'temperatur'를 'temperature'로 수정해야겠다."""
print(f"🤖 [LLM의 디버깅 사고]:\n{llm_thought}\n")

llm_code_attempt_2 = """
plt.figure(figsize=(10,5))
plt.plot(df_result['cycle'], df_result['temperature'], color='red', marker='o') # 오타 수정
plt.title("NASA Engine 1: Temperature Degradation")
plt.xlabel("Flight Cycle")
plt.ylabel("Temperature")
plt.grid(True)
plt.savefig("../engine1_report.png")
plt.close()
"""

print("⚙️ [에이전트 엔진]: 2차 수정 코드를 실행합니다...")
try:
    exec(llm_code_attempt_2)
    print("✅ [성공!] 에러 없이 실행되었으며, 그래프가 파일(../engine1_report.png)로 안전하게 저장되었습니다.")
except Exception as e:
    print(f"Error: {e}")
```
