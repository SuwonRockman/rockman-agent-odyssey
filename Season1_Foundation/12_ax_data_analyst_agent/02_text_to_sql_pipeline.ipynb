# 📊 Text-to-SQL: NASA 데이터베이스 자연어 쿼리 엔진

```python
import sqlite3
import pandas as pd
import os

db_path = './data/factory_db.sqlite'

db_schema = """
Table: engines
- engine_id (INTEGER, Primary Key)
- model_type (TEXT)

Table: sensor_logs
- engine_id (INTEGER)
- cycle (INTEGER) - 비행 횟수 (시간의 흐름)
- setting_1, setting_2, setting_3 (REAL) - 운영 환경
- sensor_1 ~ sensor_21 (REAL) - 압력, 온도 등 제트 엔진 센서
"""
print("NASA Database Schema:\n" + db_schema)
```

```python
print("👨‍💼 [지시사항]: '엔진 1번의 비행 사이클(cycle)별 센서 2번(온도)과 센서 3번(압력)의 변화를 조회해줘.'\n")

llm_generated_sql = """
SELECT cycle, sensor_2 as temperature, sensor_3 as pressure
FROM sensor_logs
WHERE engine_id = 1
ORDER BY cycle ASC;
"""
print(f"🤖 [LLM 생성 SQL]:\n{llm_generated_sql.strip()}\n")

conn = sqlite3.connect(db_path)
df_result = pd.read_sql_query(llm_generated_sql, conn)
conn.close()

print("👀 [조회 결과 Dataframe 상위 5개]:")
print(df_result.head())
```
