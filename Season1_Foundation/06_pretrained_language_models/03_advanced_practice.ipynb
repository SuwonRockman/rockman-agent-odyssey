## 1. BERT: 텍스트 감성 분류 (Fine-Tuning)
HuggingFace의 `pipeline`을 이용하여 소형 BERT 모델(DistilBERT)을 로드하고, 분류 파이프라인을 체험해 봅니다. (로컬 환경에서의 빠른 실행을 위해 이미 파인튜닝된 감성 분석 모델을 사용합니다.)

```python
from transformers import pipeline
import time

# 감성 분석(Sentiment Analysis) 파이프라인 로드
# 디폴트 모델: distilbert-base-uncased-finetuned-sst-2-english
print("BERT 감성 분석기 로딩 중...")
start = time.time()
sentiment_analyzer = pipeline("sentiment-analysis")
print(f"로딩 완료! ({time.time() - start:.2f}초)\n")

# 테스트 문장들
reviews = [
    "I absolutely loved this movie! The acting was brilliant.",
    "This was a terrible waste of time. The plot was so boring.",
    "The food was okay, but the service could be better."
]

for review in reviews:
    result = sentiment_analyzer(review)[0]
    print(f"문장: {review}")
    print(f"분석 결과: {result['label']} (확신도: {result['score']:.4f})\n")
```

## 2. GPT: 텍스트 생성 (Text Generation)
GPT 모델을 이용하여 주어진 프롬프트를 바탕으로 창의적인 텍스트를 이어서 작성하는(Autoregressive Generation) 과정을 확인해 봅니다.

```python
# 텍스트 생성 파이프라인 로드 (소형 GPT-2 모델 사용)
print("GPT-2 생성기 로딩 중...")
start = time.time()
text_generator = pipeline("text-generation", model="gpt2")
print(f"로딩 완료! ({time.time() - start:.2f}초)\n")

prompt = "Artificial Intelligence is going to"

print(f"입력 프롬프트: {prompt}")
print("-" * 50)

# max_new_tokens로 생성할 길이를 제한, num_return_sequences로 문장 개수 설정
generator_outputs = text_generator(prompt, max_new_tokens=40, num_return_sequences=2, truncation=True)

for i, output in enumerate(generator_outputs):
    print(f"[생성 결과 {i+1}]")
    print(output['generated_text'])
    print("-" * 50)
```
