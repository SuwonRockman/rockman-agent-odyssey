# 🚀 사전 학습된 임베딩: 벡터 산술(Vector Arithmetic) 분석

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

## 1. 사전 학습된 모의 Word2Vec 사운드 로드
수십기가의 GloVe를 다운로드하는 대신, 수백 차원 공간에서 단어 간의 방향성이 어떻게 성립하는지 시뮬레이션용 사전을 구축합니다.

```python
# 벡터 간의 연산을 보여주기 위한 시뮬레이션 임베딩
embeddings_dict = {
    "king": np.array([0.5, 0.9, -0.1, 0.2]),
    "man":  np.array([0.5, 0.1, -0.1, 0.1]),
    "woman": np.array([0.1, 0.1, 0.9, 0.1]),
    "queen": np.array([0.1, 0.9, 0.9, 0.2]),
    "apple": np.array([-0.9, -0.8, -0.1, 0.8]),
}

def get_cosine_sim(word1, word2):
    vec1 = embeddings_dict[word1].reshape(1, -1)
    vec2 = embeddings_dict[word2].reshape(1, -1)
    sim = cosine_similarity(vec1, vec2)[0][0]
    return sim

print(f"'king'과 'queen'의 유사도:  {get_cosine_sim('king', 'queen'):.4f}")
print(f"'king'과 'apple'의 유사도: {get_cosine_sim('king', 'apple'):.4f}")
```

## 2. 그 유명한 방정식 증명: King - Man + Woman = ?
단어가 고차원 공간의 '좌표'이자 '방향(Vector)'이 되었으므로, 덧셈 뺄셈 연산이 성립합니다.

```python
# King - Man + Woman 연산
target_vec = embeddings_dict["king"] - embeddings_dict["man"] + embeddings_dict["woman"]

# 가장 유사한 단어 찾기 알고리즘
best_word = None
best_sim = -1

for word, vec in embeddings_dict.items():
    if word in ["king", "man", "woman"]:
        continue # 입력 단어는 정답 후보에서 제외
        
    sim = cosine_similarity(target_vec.reshape(1, -1), vec.reshape(1, -1))[0][0]
    print(f"후보 '{word}' 와의 유사도: {sim:.4f}")
    
    if sim > best_sim:
        best_sim = sim
        best_word = word

print(f"\n🎯 방정식의 정답: '{best_word}' (유사도 {best_sim:.4f})")
print("결론: 임베딩 벡터 공간 안에서 '성별의 차이'를 가리키는 방향 벡터가 완벽하게 수학적으로 존재함을 증명합니다.")
```
